```python
import numpy as np
from iminuit import Minuit
from iminuit.cost import LeastSquares
from scipy.stats import chi2
import matplotlib.pyplot as plt

def line(x, alpha, beta):
    return beta*x +alpha

data_x = [0.05895, 0.11786, 0.17669, 0.23562, 0.29481] #correnti ampere(errori ampere trascurabili, ordine 0.001 milliAmpere)
data_y = [2.379, 4.754, 7.130, 9.500, 11.890] #tensioni volt
 
y_err = [0.121, 0.240, 0.377, 0.495, 0.615] #errore tensioni


least_squares = LeastSquares(data_x, data_y, y_err, line)
m = Minuit(least_squares, alpha=0, beta=0)  # initial parameters

m.migrad()  # finds minimum of least_squares function
m.hesse()  # accurately computes uncertaintiesù
print (m.valid)
display (m)
print ('Value of Q2: ', m.fval)
print ('Number of degrees of freedom: ', m.ndof)

print ('associated p-value: ', 1. - chi2.cdf (m.fval, df = m.ndof))
print('coefficinete R = ', m.values['beta'], 'errore = ', m.errors['beta'])
print('intercetta = ', m.values['alpha'], 'errore = ', m.errors['alpha'])

fig, ax = plt.subplots()

ax.errorbar(data_x, data_y, y_err, fmt="o", label = 'dati sperimentali')
plt.plot(data_x, line(np.array(data_x), *m.values), label = 'Configurazione 2: 40 Ohm')
plt.xlabel("Corrente [A]")
plt.ylabel("Tensione [V]")

plt.legend()
plt.show()

```
